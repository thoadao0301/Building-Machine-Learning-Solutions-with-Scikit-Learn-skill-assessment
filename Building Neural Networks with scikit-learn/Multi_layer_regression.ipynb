```python
from sklearn.model_selection import train_test_split

X = diet_df.drop(['change','Date','Stone','Pounds','Ounces'],axis=1)
y = diet_df['change']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp_reg = MLPRegressor(activation='relu',hidden_layer_sizes=(1,),solver='lbfgs',verbose=True,max_iter=100)
```

```python
mlp_reg = MLPRegressor(activation='relu',hidden_layer_sizes=(5,),solver='lbfgs',verbose=True,max_iter=100)
```

```python
mlp_reg = MLPRegressor(activation='relu',hidden_layer_sizes=(2,4),solver='lbfgs',verbose=True,max_iter=100)
```

```python
mlp_reg = MLPRegressor(activation='logistic',hidden_layer_sizes=(1,),solver='lbfgs',verbose=True,max_iter=100)
```

```python
mlp_reg.fit(X_train,y_train)
```

```python
y_pred = mlp_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print('Trainng score: ',mlp_reg.score(X_train,y_train))
print('Testing score: ',r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(20,10))

plt.plot(y_pred,label='Predictions')
plt.plot(y_test.values,label='Actual')

plt.ylabel('Change in weight')

plt.legend()
plt.show()
```
