```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.metrics import silhouette_score
from sklearn import metrics

from sklearn.model_selection import ParameterGrid

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import MeanShift
```

```python
path_data = 'C:/Users/baoth/Desktop/Learnmore/DS/machine_learning_PS/data/data_1024.csv'
```

```python
df = pd.read_csv(path_data,sep='\t',header=None)
```

```python
df.columns = ['Driver_ID','Distance_Feature','Speeding_Feature']
df.head()
```

```python
df.shape
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.describe()
```

```python
features = df.drop('Driver_ID',axis=1)
```

```python
parameters = {'n_clusters' : [2,3,4,5,10,20,30]}

parameters_grid = ParameterGrid(parameters)
```

```python
list(parameters_grid)
```

```python
best_score = -1
model = KMeans()
```

```python
for g in parameters_grid:
    model.set_params(**g)
    model.fit(features)

    ss = metrics.silhouette_score(features,model.labels_)
    print('Parameter: ', g, 'Score: ',ss)
    if ss > best_score:
        best_score = ss
        best_grid = g
```

```python
best_grid
```

```python
fig,ax = plt.subplots(figsize=(12,8))

plt.scatter(features['Distance_Feature'],
            features['Speeding_Feature'],s=250)

plt.xlabel('Distance_Feature')
plt.ylabel('Speeding_Feature')
plt.show()
```

```python
parameters = {'eps' : [0.9,1.0,5.0,10.0,12.0,14.0,20.0],
            'min_samples': [5,7,10,12]}

parameters_grid = ParameterGrid(parameters)
list(parameters_grid)
```

```python
model = DBSCAN()
best_score = -1
```

```python
for g in parameters_grid:
    model.set_params(**g)
    model.fit(features)

    ss = metrics.silhouette_score(features,model.labels_)
    print('Parameter: ', g, 'Score: ', ss)
    if ss > best_score:
        best_score = ss
        best_grid = g
```

```python
best_grid
```

```python
model.set_params(**best_grid)
model.fit(features)
```

```python
len(model.labels_)
```

```python
n_cluster = len(set(model.labels_)) - (1 if -1 in model.labels_ else 0)
n_cluster
```

```python
n_noise = list(model.labels_).count(-1)
n_noise
```

```python
another_grid = {'eps' : 5.0, 'min_samples' : 5}

model.set_params(**another_grid)
model.fit(features)

n_cluster = len(set(model.labels_)) - (1 if -1 in model.labels_ else 0)
print('Number of Clusters: ',n_cluster)

n_noise = list(model.labels_).count(-1)
print('Noisy points: ',n_noise)
```

```python
from sklearn.cluster import estimate_bandwidth
estimate_bandwidth(features)
```

```python
model = MeanShift(bandwidth=estimate_bandwidth(features)).fit(features)
metrics.silhouette_score(features,model.labels_)
```
